# k-means

## k-meansの仕組み
---
クラスタ重心 (クラスタ内の学習サンプル平均) からの距離でデータを k 個に分割。
1. クラスタ中心を適当に選ぶ
1. データ点を一番距離の近いクラスタ中心に割り当てる
1. 割り当てられたデータ点に基づいてクラスタ中心を計算し直す
1. 2~3 を収束するまで繰り返す
